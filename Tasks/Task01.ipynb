```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_87.csv', skiprows=4)
```

```python
df_2023 = df[['Country Name', '2023']].dropna()
df_2023['2023'] = pd.to_numeric(df_2023['2023'], errors='coerce')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_2023['2023'], bins=30, color='skyblue', edgecolor='black', log=True)  

plt.xlabel('Population (log scale)')
plt.ylabel('Number of Countries')
plt.title('Distribution of Total Population Across Countries (2023)')

plt.grid(True)
plt.show()
```
